```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
df.sample(3)
```

```python
df['pci']=df['per capita income (US$)']
df.head(3)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df.pci,color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.pci)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
